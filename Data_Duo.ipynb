```python
import pandas as pd

# Load the Excel file
file_path = "C:/Users/user/Desktop/Data_Duo/EC 2023 Tables_Report_Final 0n 9 july 2024_.xlsx"

# Read all sheet names
try:
    sheet_names = pd.ExcelFile(file_path).sheet_names
    print("Available sheets:", sheet_names)
except Exception as e:
    print(f"Error loading sheets: {e}")
    exit()

# Dictionary to store cleaned data
cleaned_sheets = {}

# Iterate through all sheets
for sheet in sheet_names:
    print(f"\nProcessing sheet: {sheet}")
    try:
        # Load the sheet into a DataFrame
        df = pd.read_excel(file_path, sheet_name=sheet, engine='openpyxl')
        print(f"Loaded sheet '{sheet}' successfully!")

        # Clean the data
        # Drop rows and columns that are completely empty
        df = df.dropna(how="all", axis=0).dropna(how="all", axis=1)

        # Reset index for clean formatting
        df = df.reset_index(drop=True)

        # Assign default headers if necessary
        if df.columns.isnull().any():
            df.columns = [f"Column{i+1}" for i in range(df.shape[1])]

        # Store cleaned DataFrame
        cleaned_sheets[sheet] = df

    except Exception as e:
        print(f"Error processing sheet '{sheet}': {e}")
        continue

# Save all cleaned sheets to a new Excel file
output_file = "C:/Users/user/Desktop/Data_Duo/cleaned_all_sheets.xlsx"
try:
    with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
        for sheet_name, cleaned_df in cleaned_sheets.items():
            cleaned_df.to_excel(writer, index=False, sheet_name=sheet_name)
    print(f"\nAll cleaned sheets saved to {output_file}")
except Exception as e:
    print(f"Error saving cleaned data: {e}")
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load Table 2.1.4
file_path = "C:/Users/user/Desktop/Data_Duo/EC 2023 Tables_Report_Final 0n 9 july 2024_.xlsx"
sheet_name = "Table 2.1.4"

# Read the data
df4 = pd.read_excel(file_path, sheet_name=sheet_name, engine="openpyxl")

# Drop completely empty rows and columns
df4 = df4.dropna(how="all", axis=0).dropna(how="all", axis=1)

# Reset index after dropping
df4.reset_index(drop=True, inplace=True)

# Rename columns for clarity
df4.columns = ["District/Province", "2014 Count", "2017 Count", "2020 Count", "2023 Count", 
               "2014 Percentage", "2017 Percentage", "2020 Percentage", "2023 Percentage"]

# Drop rows that are not meaningful (e.g., NaN or total rows)
df4 = df4[~df4["District/Province"].str.contains("NaN|Total", na=False)]

# Convert numeric columns to numeric type
numeric_cols = ["2014 Count", "2017 Count", "2020 Count", "2023 Count", 
                "2014 Percentage", "2017 Percentage", "2020 Percentage", "2023 Percentage"]
df4[numeric_cols] = df4[numeric_cols].apply(pd.to_numeric, errors="coerce")

# Handle missing values in 'District/Province' column (e.g., fill NaN with 'Unknown')
df4["District/Province"] = df4["District/Province"].fillna("Unknown").astype(str)

# Display cleaned data
print("Cleaned Table 2.1.4:")
print(df4.head())

# Create a new column for percentage change from 2020 to 2023
df4["Change 2020-2023 (%)"] = ((df4["2023 Count"] - df4["2020 Count"]) / df4["2020 Count"]) * 100

# Visualization: Bar plot for the change in establishments from 2020 to 2023
plt.figure(figsize=(12, 6))
plt.barh(df4["District/Province"], df4["Change 2020-2023 (%)"], color="lightcoral")

plt.title("Percentage Change in Establishments (2020 to 2023) by District/Province", fontsize=16)
plt.xlabel("Percentage Change")
plt.ylabel("District/Province")
plt.tight_layout()

# Save and show the plot
plt.savefig("images/percentage_change_establishments_2020_2023.png")
plt.show()



```
